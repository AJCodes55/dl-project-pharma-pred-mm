# Clinical Trial Data Extraction (Phase Extension)

This notebook helps build `clinical_trial_event_calendar.csv` from ClinicalTrials.gov CSV exports.

## Input
Place downloaded CSV files in `raw_clinical_trials/`.

## Output (in `processed/`)
- `clinical_trial_event_calendar.csv`
- `clinical_trial_studies_clean.csv`
- `clinical_trial_summary.json`


```python
# Optional dependencies (usually already available)
# %pip install pandas
```

```python
from pathlib import Path
import pandas as pd
import json

from extract_clinical_trials import run_pipeline
```

```python
RAW_DIR = Path("raw_clinical_trials")
OUT_DIR = Path("processed")

print("RAW_DIR exists:", RAW_DIR.exists())
if RAW_DIR.exists():
    files = list(RAW_DIR.glob("*.csv"))
    print("CSV files found:", len(files))
    for f in files[:10]:
        print(" -", f.name)
```

```python
# Run extraction
run_pipeline(
    raw_dir=RAW_DIR,
    out_dir=OUT_DIR,
    min_event_date="2012-01-01",
    max_event_date="2023-12-31",
    keep_statuses=None,
)
```

```python
# Inspect outputs
events = pd.read_csv("processed/clinical_trial_event_calendar.csv", parse_dates=["date"])
clean = pd.read_csv("processed/clinical_trial_studies_clean.csv")
summary = json.loads(Path("processed/clinical_trial_summary.json").read_text())

print("events shape:", events.shape)
print("clean studies shape:", clean.shape)
print("
summary:")
print(json.dumps(summary, indent=2))
```

```python
# Quick report tables
print("
Events per ticker:")
print(events["ticker"].value_counts().sort_index())

print("
Events per type:")
print(events["event_type"].value_counts().sort_values(ascending=False).head(20))

events.head(10)
```
