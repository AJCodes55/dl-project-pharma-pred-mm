```python
# Inspect key outputs for all ablations
from pathlib import Path
import pandas as pd

base = Path('results')
paths = {
    'basic_metrics': base / 'phase5_basic_fda' / 'phase5_fda_metrics.csv',
    'rich_metrics': base / 'phase5_rich_fda' / 'phase5_fda_metrics.csv',
    'rich_ct_metrics': base / 'phase5_rich_fda_ct' / 'phase5_fda_metrics.csv',
    'basic_trials': base / 'phase5_basic_fda' / 'phase5_robust_search_trials.csv',
    'rich_trials': base / 'phase5_rich_fda' / 'phase5_robust_search_trials.csv',
    'rich_ct_trials': base / 'phase5_rich_fda_ct' / 'phase5_robust_search_trials.csv',
    'basic_comp': base / 'phase5_basic_fda' / 'phase3_phase4_phase5_comparison.csv',
    'rich_comp': base / 'phase5_rich_fda' / 'phase3_phase4_phase5_comparison.csv',
    'rich_ct_comp': base / 'phase5_rich_fda_ct' / 'phase3_phase4_phase5_comparison.csv',
    'basic_interp': base / 'phase5_basic_fda' / 'phase5_fda_interpretability.csv',
    'rich_interp': base / 'phase5_rich_fda' / 'phase5_fda_interpretability.csv',
    'rich_ct_interp': base / 'phase5_rich_fda_ct' / 'phase5_fda_interpretability.csv',
}

for name, p in paths.items():
    print(f'{name}:', 'OK' if p.exists() else 'MISSING', '-', p)

if paths['basic_metrics'].exists():
    print('\nBasic FDA metrics:')
    display(pd.read_csv(paths['basic_metrics']))
if paths['rich_metrics'].exists():
    print('\nRich FDA metrics:')
    display(pd.read_csv(paths['rich_metrics']))
if paths['rich_ct_metrics'].exists():
    print('\nRich FDA + CT metrics:')
    display(pd.read_csv(paths['rich_ct_metrics']))
```
